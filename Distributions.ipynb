```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dt = pd.read_csv('IRIS.csv')
```

```python
dt.head()
```

```python
sns.distplot(a=dt['petal_length'], kde=False)
```

```python
# KDE plot 
sns.kdeplot(data=dt['petal_length'], shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=dt['petal_length'], y=dt['sepal_width'], kind="kde")
```
